```python
popuu_1998=df['POPULATION 1998 (URBAN)'].sum()
popur_1998=df['POPULATION 1998 (RURAL)'].sum()
popu_1998=popur_1998 + popuu_1998
popur_2017=df['ALL SEXES (RURAL)'].sum()
popuu_2017=df['ALL SEXES (URBAN)'].sum()
popu_2017=popur_2017+popuu_2017
print(popu_2017)
print(popu_1998)
```

```python
print('The total population of Pakistan was increased  by 2017 as compared to 1998 is : ',popu_2017-popu_1998)
print('The total population of Pakistan was increased  by 2017 as compared to 1998 in percentage : ',(popu_1998)/popu_2017*100)
```

```python
df[['POPULATION 1998 (RURAL)','POPULATION 1998 (URBAN)']].sum()
```

```python

df['Total_1998'] = df['POPULATION 1998 (URBAN)'] + df['POPULATION 1998 (RURAL)']
df['Total_2017'] = df['ALL SEXES (URBAN)'] + df['ALL SEXES (RURAL)']

data = df[['Total_1998', 'Total_2017']].sum().reset_index()
data.columns = ['Year', 'Population']
data['Year'] = ['1998', '2017']

sns.barplot(data=data, x='Year', y='Population')
plt.title("Population Comparison")
plt.show()
```
