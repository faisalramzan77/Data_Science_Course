# Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
```

```python
df.plot(kind='scatter', x='age', y='fare', title='Scatter Plot', color='blue')
```

```python
# Change above graph color using Hex Code
df.plot(kind='scatter', x='age', y='fare', title='Scatter Plot', color='#5C1B1B')
```

```python
df.head()
```

```python
df.groupby('sex')['age'].mean()
```

```python
df.groupby('sex')['age'].mean().plot(kind='bar')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
```

```python
df.groupby('sex')['fare'].mean().plot(kind='bar')
plt.title('Average Fare by Sex')
plt.ylabel('Average Fare')
plt.xlabel('Gender')
```

```python
#Bar plot usnig seaborn libraries
sns.barplot(x="survived", y="fare", data=df)
```

```python
#Bar plot usnig seaborn libraries
sns.barplot(x="survived", y="fare", data=df, hue="class")
```

```python
# categorical
sns.boxplot(x="survived", y="age", data=df, hue="class")
```

```python
# numeric
sns.scatterplot(x="fare", y="age", data=df, hue="class")
```

```python
sns.scatterplot(y="fare", x="age", data=df, hue="class")
```

```python
# make scatter plot via plotly
import plotly.express as px
fig = px.scatter(df, x='age', y='fare', color='sex')
fig.show()
```

```python
import plotly.graph_objects as go

import pandas as pd

# Read data from a csv
z_data = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/api_docs/mt_bruno_elevation.csv')

fig = go.Figure(data=[go.Surface(z=z_data.values)])

fig.update_layout(title=dict(text='Mt Bruno Elevation'), autosize=False,
                  width=500, height=500,
                  margin=dict(l=65, r=50, b=65, t=90))

fig.show()
```
