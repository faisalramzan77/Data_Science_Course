```python
# Data
df = sns.load_dataset('titanic')
```

```python
df.info
```

```python
# Scatter plt
# scatter plot size
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# Scatter plt
# scatter plot size
plt.figure(figsize=(16,9))
sns.scatterplot(data=df, x='age', y='fare', hue='sex')
```

```python
# Scatter plt
# scatter plot size
plt.figure(figsize=(16,9))
palette = {'male':'green', 'female':'red'}
sns.scatterplot(data=df, x='age', y='fare', hue='sex', palette=palette)
# change the colors
plt.title('Scatter plot of age vs fare with sex as hue')

# saving a plot 300dpi, 600dpi, 1500dpi png 
plt.savefig('../output/plots/sscatter_plot.svg', dpi=300,)
```

```python
# Scatter plt
# scatter plot size
plt.figure(figsize=(16,9))
palette = {'First':'green', 'Second':'blue', 'Third':'red', 'Fourth':'yellow'}
sns.boxplot(data=df, x='sex', y='age', hue='class', palette=palette)
# change the colors
plt.title('Scatter plot of age vs fare with sex as hue')

# saving a plot 300dpi, 600dpi, 1500dpi png 
plt.savefig('../output/plots/boxx_plot.png', dpi=300,)
```

```python
# Scatter plt
# scatter plot size
plt.figure(figsize=(16,9))
palette = {'First':'green', 'Second':'blue', 'Third':'red', 'Fourth':'yellow'}
sns.boxenplot(data=df, x='sex', y='age', hue='class', palette=palette)
# change the colors
plt.title('Scatter plot of age vs fare with sex as hue')

# saving a plot 300dpi, 600dpi, 1500dpi png 
plt.savefig('../output/plots/boxxen_plot.png', dpi=300,)
```
