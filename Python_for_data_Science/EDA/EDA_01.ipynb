```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('../Data/titanic.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('deck', axis=1, inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
df['age'].fillna(df['age'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)
df.isnull().sum()
```

```python
# Heatmap to visualize missing values
sns.heatmap(df.isnull())
```

# Feature Engineering

```python
# Binnning of age column into 7 categories
bins = [0, 1, 5, 12, 18 ,30, 50 ,80]
lables = ['infants', 'Toddlers', 'Kids', 'Teens', 'Youngs', 'middle aged', 'Old']
# Whos columns we want to bin
df['binned_age'] = pd.cut(df['age'], bins=bins, labels=lables)
```

```python
df.head()
```

```python
df['binned_age'].value_counts()
```
