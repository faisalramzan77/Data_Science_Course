# Split a string into multiple columns

```python
import pandas as pd
df = pd.DataFrame({'Name': ['Faisal Ramzan', 'Sultan Ramzan', 'Usman Ali'], 'Age': [25, 30, 35], 'Address':['Faisalabad Pakistan','Multan Pakistan','Lahore Pakistan']}) # Create a DataFrame with two columns
df
```

```python
df[['F_Name','L_Name']] = df.Name.str.split(' ', expand=True) # Split the 'name' column into multiple columns
```

```python
df
```

```python
df[['city', 'country']] = df.Address.str.split(' ', expand=True) # Split the 'Address' column into multiple columns
df
```

```python
# refine data manipulation
df  = df[['F_Name', 'L_Name', 'city', 'country']] # Refine the DataFrame to include only specific columns
df
```
