```python
# How to rename columns in a DataFrame
df.rename(columns={'a Col': 'A_Col', 'b Col': 'B_Col'}, inplace=True) # Rename columns in the DataFrame
df
```

```python
# rename colums in another way
df.columns = ['A_Colll', 'B_Colll'] # Rename columns by directly assigning a new list of column names
df
```

```python
# rename just string columns or columns that contain a specific string
df.columns = df.columns.str.replace('lll', 'l') # Replace 'Col' with 'Column' in all column names
df
```

```python
# adding prefix to columns
df = df.add_prefix('new_') # Add a prefix to all column names
df
```

```python
# adding suffix to columns
df = df.add_suffix('end') # Add a suffix to all column names
df
```

```python
# Reverse Row Order
import pandas as pd
df = pd.read_csv("../Data/titanic.csv") # Read the DataFrame from the CSV file
print(df)
print("Below is the reverse of the DataFrame")
df.loc[::-1].head() # Reverse the order of rows in the DataFrame
```
